# **My First ML Project**

## **Load Data**

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
df
```

## **Data Preparation**

### **Data Split into X and y**



```python
y = df['logS']
y
```

#### **To view the data in columns for drop axis=1 and rows for axis=0**

```python
X = df.drop ( 'logS', axis=1)
X
```

### **Data Splitting into training data and testing data**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=100)
```

#### **Hence, the train set will have 80% of the data and the test set will have 20%**

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

## **Model Building**



### **Linear Regression**

#### **Training the Model**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
```

#### **Applying the model to make prediction**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred
```

```python
y_lr_test_pred
```

## **Evaluate Model Performance**

```python
y_train
```

```python
y_lr_train_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print ('LR MSE (train)',lr_train_mse )
print ('LR R2 (train)',lr_train_r2 )
print ('LR MSE (test)',lr_test_mse )
print ('LR R2 (test)',lr_test_r2 )
```

```python
lr_results = pd.DataFrame(['LinearRegression()', lr_train_mse, lr_train_r2, lr_test_mse,  lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Testing MSE', 'Testing R2']
```

```python
lr_test_r2
```

```python
lr_results
```

## **Random Forest**


### **Training the Model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### **Applying the model to make prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

#### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print ('RF MSE (train)',rf_train_mse )
print ('RF R2 (train)',rf_train_r2 )
print ('RF MSE (test)',rf_test_mse )
print ('RF R2 (test)',rf_test_r2 )
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2] ).transpose()
rf_results.columns = ['Method','Training MSE', 'Training R2', 'Testing MSE', 'Testing R2']
rf_results
```

## **Model Comparison**



```python
df_models = pd.concat([lr_results, rf_results], axis=0)
```

## **Data Visualization of Prediction Results**


```python
import matplotlib.pyplot as plt

plt.scatter(x = y_train, y = y_lr_train_pred)
plt.ylabel('Predict Logs')
plt.xlabel('Experimental Logs')
plt.title('Linear Regression')
plt.plot()
```

```python
import numpy as np

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)

plt.scatter(x = y_train, y = y_lr_train_pred)
plt.ylabel('Predict Logs')
plt.xlabel('Experimental Logs')
plt.title('Linear Regression')
plt.plot(y_train, p(y_train), '#F8766D')
```
